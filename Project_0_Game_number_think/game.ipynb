# Game 'Think Numbers'

The game consists of one algorithm guessing a number. And the second algorithm guesses it. 

It was interesting to see how the number of attempts depends on the number itself. 

Below is the code that runs the functions from the game a certain number of times to see the average number of attempts.

```python
from game import mean_counts
mean_arr = []
for i in range(1000):
    mean_arr.append(mean_counts(i))
import numpy as np
import matplotlib.pyplot as plt

plt.plot(mean_arr)
plt.show()
```
